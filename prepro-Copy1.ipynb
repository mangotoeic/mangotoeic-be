```python
import pandas as pd
data_path = './data/train.csv'
df = pd.read_csv(data_path)
df
```

```python
df1=df.drop(['row_id', 'timestamp', 'content_type_id',
       'task_container_id', 
       'prior_question_elapsed_time'], axis=1)
df1
```
